# 3. Average grade

```python
grades = grades.with_column('Average Grades', grades.column('math score')*4/10 + grades.column('reading score')*3/10 + grades.column('writing score')*3/10)
grades.drop('math score', 'reading score', 'writing score').show()
# grades
```

```python
grades = grades.sort('Average Grades', descending = True)
grades.where('Average Grades', are.above(95)).num_rows
```

```python
grades.join('ID', background, 'ID').drop('ID', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course').group('gender', np.average)
# The males only did better in maths on average, the other scores are higher for females
```

# 7. Letter Grade

```python
def let_grade(grade):
    if grade >= 90:
        return 'A'
    if 80 <= grade <90:
        return 'B'
    if 70 <= grade <80:
        return 'C'
    if 60 <= grade <70:
        return 'D'
    else:
        return 'F'
```

```python
vals = grades.apply(let_grade, 'Average Grades')
grades = grades.with_column('Letter Grade', vals)
# grades.drop('ID', 'math score', 'reading score', 'writing score').plot('Letter Grade')
grades.group('Letter Grade').plot('Letter Grade')
```

```python
grades.group('Letter Grade').sort('count', descending = True).take(0)
```

# 8. Race/Ethnic Group Analysis

```python
grades.join('ID', background, 'ID').where('Letter Grade', 'F').group('race/ethnicity').sort('count', descending=True).take(0)
```

# 10. Course Preparation Distributions

```python
my_bin = make_array(0,50,60,65,70,75,80,90,100)
none = background.join('ID', grades, 'ID').where('test preparation course', 'none')
comp = background.join('ID', grades, 'ID').where('test preparation course', 'completed')

none.hist('math score', bins = my_bin), comp.hist('math score', bins = my_bin)
```

```python
none.hist('reading score', bins = my_bin), comp.hist('reading score', bins = my_bin)
```

```python
none.hist('writing score', bins = my_bin), comp.hist('writing score', bins = my_bin)
```

# 11. Education Level

```python
num_col = grades.join('ID', background, 'ID').where('parental level of education', are.not_containing('high school'))
num_all = grades.join('ID', background, 'ID')
num_col_ab = num_col.where('Letter Grade', are.below_or_equal_to('B')).num_rows
num_all_ab = num_all.where('Letter Grade', are.below_or_equal_to('B')).num_rows
per_col = num_col_ab/num_col.num_rows
per_all = num_all_ab/num_all.num_rows
per_col, per_all
# num_col
```
